```python
import pandas as pd
from sklearn.calibration import calibration_curve
import matplotlib.pyplot as plt
import matplotlib.lines as mlines
```

```python
data = pd.read_pickle('best_model/evaluate_validation.pkl')
data = pd.DataFrame(data)
data
```

```python
prediction_ = data["probs"].tolist()
prediction = []
for i in range(len(prediction_)):
    prediction.append(prediction_[i][1])
```

```python
y, x = calibration_curve(data["labels"].tolist(), prediction, n_bins=20)
```

```python
fig, ax = plt.subplots()

# reference line, legends, and axis labels
line = mlines.Line2D([0, 1], [0, 1], color='tab:blue', label="Ground truth")
# only these two lines are calibration curves
plt.plot(x,y, marker='o', linewidth=1, color="tab:orange", label='Our model')

transform = ax.transAxes
line.set_transform(transform)
ax.add_line(line)
fig.suptitle('Calibration plot for validation data')
ax.set_xlabel('Predicted probability')
ax.set_ylabel('True probability in each bin')
plt.legend()
plt.show()
```
