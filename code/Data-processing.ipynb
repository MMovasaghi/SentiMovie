## Embedding representation

```python
from sklearn.manifold import TSNE
import umap
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
```

```python
all_sample = pd.read_json("./../data/sample_embedding.json")
all_sample['review_content_bert_embedding'] = [np.asarray([float(x) for x in t[1:-1].split()]) 
                                                   for t in all_sample['review_content_bert_embedding']]
all_sample
```

```python
def plot_data(df, title, axis_name):
    plt.figure(figsize=(6,4))

    rndperm = np.random.permutation(df.shape[0])

    sns.scatterplot(
        x=axis_name[0], y=axis_name[1],
        hue="label",
        palette=sns.color_palette("hls", len(np.unique(df['label']))),
        data=df.loc[rndperm,:],
        legend="full",
        alpha=0.9
    )
    plt.title(title)
    plt.show()
```

```python
X_embedded_bert = []
for i in all_sample['review_content_bert_embedding']:
    X_embedded_bert.append(np.asarray(i))
X_embedded_bert = np.array(X_embedded_bert)

X_embedded_tfidf = []
for i in all_sample['review_content_tf_idf_embedding']:
    X_embedded_tfidf.append(np.asarray(i))
X_embedded_tfidf = np.array(X_embedded_tfidf)

X_embedded_bert.shape, X_embedded_tfidf.shape
```

```python
X_embedded_bert_tsne = TSNE(n_components=2, 
                            learning_rate='auto', 
                            init='random', 
                            random_state=42, 
                            perplexity=3).fit_transform(X_embedded_bert)

X_embedded_tfidf_tsne = TSNE(n_components=2, 
                            learning_rate='auto', 
                            init='random', 
                            random_state=42, 
                            perplexity=3).fit_transform(X_embedded_tfidf)

X_embedded_bert_umap = umap.UMAP(n_components=2, 
                                   random_state=42).fit_transform(X_embedded_bert)

X_embedded_tfidf_umap = umap.UMAP(n_components=2, 
                                   random_state=42).fit_transform(X_embedded_tfidf)

df = {"bert_tsne-1": X_embedded_bert_tsne[:,0], 
       "bert_tsne-2": X_embedded_bert_tsne[:,1], 
      "bert_umap-1": X_embedded_bert_umap[:,0], 
       "bert_umap-2": X_embedded_bert_umap[:,1],
       "tfidf_tsne-1": X_embedded_tfidf_tsne[:,0], 
       "tfidf_tsne-2": X_embedded_tfidf_tsne[:,1], 
      "tfidf_umap-1": X_embedded_tfidf_umap[:,0], 
       "tfidf_umap-2": X_embedded_tfidf_umap[:,1],
       "label": list(all_sample['review_type'])}
df = pd.DataFrame(df)
df
```

```python
plot_data(df, "UMAP-BERT", ['bert_umap-1', 'bert_umap-2'])
plot_data(df, "TSNE-BERT", ['bert_tsne-1', 'bert_tsne-2'])

plot_data(df, "UMAP-Tf-idf", ['tfidf_umap-1', 'tfidf_umap-2'])
plot_data(df, "TSNE-Tf-idf", ['tfidf_tsne-1', 'tfidf_tsne-2'])
```

## Finding mapping function for movie score

```python
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
eda.movies = pd.read_csv("./../data/normalized_movie.csv")
eda.movies
```

```python
movie_tmp = eda.movies.copy()
movie_tmp['diff_tom_aud'] = [t-a for t,a in zip(movie_tmp['tomatometer_rating'], movie_tmp['audience_rating'])]
plt.hist(movie_tmp['diff_tom_aud'], bins=100, label="Data")
plt.plot([-50]*48000, np.linspace(0,48000, 48000), label="Threshold-left-tail")
plt.plot([50]*48000, np.linspace(0,48000, 48000), label="Threshold-right-tail")
plt.title("Different rating distribution")
plt.xlabel('differences')
plt.ylabel('Number')
plt.legend()
plt.show()
```

```python
left_tail = movie_tmp[(movie_tmp['diff_tom_aud'] < -50)].sample(30)
right_tail = movie_tmp[(movie_tmp['diff_tom_aud'] > 50)].sample(30)
middle = movie_tmp[(movie_tmp['diff_tom_aud'] < 50) & (movie_tmp['diff_tom_aud'] > -50)].sample(40)
check_frame = pd.concat([left_tail, right_tail, middle])
eda.save_data(check_frame, path="./../data/sampled_data_for_score_finding.csv")
check_frame
```

```python
df = pd.read_csv('./../data/sampled_data_with_IMDB_Score.csv')
```

```python
#Setting the value for X and Y
x = df[['tomatometer_rating', 'audience_rating']]/100
y = df['IMDB_score']/10

#Splitting the dataset
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=42)

#Fitting the Multiple Linear Regression model
mlr = LinearRegression()  
model = mlr.fit(x_train, y_train)

print(f"{model.intercept_:.6f} + ({model.coef_[0]:.6f} * tomatometer_rating) + ({model.coef_[1]:.6f} * audience_rating)")
```

```python
def my_score_mapping(tomatometer_rating, audience_rating, model_param=[0, 0]):
    return (model_param[0] * tomatometer_rating) + (model_param[1] * audience_rating)
```

### Evaluation

```python
y_pred_mlr= model.predict(x_test)
slr_diff = pd.DataFrame({'Actual value': y_test, 'Predicted value': y_pred_mlr})
slr_diff
```

```python
meanAbErr = metrics.mean_absolute_error(y_test, y_pred_mlr)
meanSqErr = metrics.mean_squared_error(y_test, y_pred_mlr)
rootMeanSqErr = np.sqrt(metrics.mean_squared_error(y_test, y_pred_mlr))
print('R squared: {:.2f}'.format(model.score(x,y)*100))
print('Mean Absolute Error:', meanAbErr)
print('Mean Square Error:', meanSqErr)
print('Root Mean Square Error:', rootMeanSqErr)
```

### Apply score function on movie data

```python
eda.movies = pd.read_csv('./../data/normalized_movie.csv', index_col=False)
```

```python
eda.movies['my_score'] = [my_score_mapping(t, a, model_param=[model.coef_[0], model.coef_[1]]) 
                          for t, a in zip(eda.movies['tomatometer_rating'], eda.movies['audience_rating'])]
eda.movies = eda.movies[['rotten_tomatoes_link', 'movie_title', 
                         'review_score', 'review_type', 
                         'review_content', 'my_score']].copy()
eda.movies
```
