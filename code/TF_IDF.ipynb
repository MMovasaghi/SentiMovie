```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tqdm import tqdm
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
import nltk
from sklearn.feature_extraction.text import TfidfVectorizer
import re
pd.options.mode.chained_assignment = None
```

```python
class TF_IDF:
    df = None
    coulmn_name = "tf_idf"
    ngram_range = (2,7)
    data_type = np.float32
    def __init__(self, df):
        self.df = df
        
    def perform_tf_idf(self, coulmn_name):
        vectorizer = TfidfVectorizer(analyzer="char",
            ngram_range=self.ngram_range,
            dtype=self.data_type
        )
        return vectorizer.fit_transform(self.df[coulmn_name])

    def add_coulmn_to_df(self, tf_idf_vector):
        self.df[self.coulmn_name] = None
        for index, row in self.df.iterrows():
            self.df[self.coulmn_name][index] = tf_idf_vector[index]
    
    def save_data(self, path):
        self.df.to_csv(path, index=False)
```

```python
pos_reviews = pd.read_csv("./../data/fresh_sample_embedding.csv")
neg_reviews = pd.read_csv("./../data/rotten_sample_embedding.csv")
```

```python
pos_tf_idf = TF_IDF(pos_reviews)
pos_tf_idf.add_coulmn_to_df(pos_tf_idf.perform_tf_idf('review_content').toarray())
pos_tf_idf.save_data("./../data/fresh_sample_embedding_with_tf_idf.csv")
```

```python
neg_tf_idf = TF_IDF(neg_reviews)
neg_tf_idf.add_coulmn_to_df(neg_tf_idf.perform_tf_idf('review_content').toarray())
neg_tf_idf.save_data("./../data/rotten_sample_embedding_with_tf_idf.csv")
```
